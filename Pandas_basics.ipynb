```python
# add a new column to the DataFrame containing only the month of the measurement

air_quality_no2_long["month"] = air_quality_no2_long["datetime"].dt.month
air_quality_no2_long.head()
```

```python
# By using Timestamp objects for dates, a lot of time-related properties are provided by pandas. For example the month, but also year, weekofyear, quarter,… 
# All of these properties are accessible by the dt accessor.
```

```python
air_quality_no2_long["year"] = air_quality_no2_long["datetime"].dt.year
air_quality_no2_long.head()
```

```python
# The average NO2 concentration for each day of the week for each of the measurement location:

air_quality_no2_long.groupby([air_quality_no2_long["datetime"].dt.weekday,"location"])["value"].mean()
```

```python
# NOTE: As we are working with a very short time series in these examples, the analysis does not provide a long-term representative result!
```

```python
fig, axs = plt.subplots(figsize = (12,4))
air_quality_no2_long.groupby(air_quality_no2_long["datetime"].dt.hour)["value"].mean().plot(kind = 'bar', rot = 0, ax=axs)
plt.xlabel("Hour of the day")
plt.ylabel("$NO_2 (µg/m^3)$")
```

```python
NO2 = air_quality_no2_long.pivot(index="datetime",columns="location",values="value")
NO2.head()
```

```python
# Working with a datetime index (i.e. DatetimeIndex) provides powerful functionalities. For example, we do not need the dt accessor to get the time series properties, but have these properties available on the index directly:

NO2.index.year,NO2.index.weekday
```

```python
# Create a plot of the NO2 values in the different stations from the 20th of May till the end of 21st of May

NO2["2019-05-20":"2019-05-21"].plot()
```

```python
# Aggregate the current hourly time series values to the monthly maximum value in each of the stations.

monthly_max = NO2.resample("M").max()
monthly_max
```

```python
# A very powerful method on time series data with a datetime index, is the ability to resample() time series to another frequency (e.g., converting secondly data into 5-minutely data).

# The resample() method is similar to a groupby operation:

# it provides a time-based grouping, by using a string (e.g. M, 5H,…) that defines the target frequency

# it requires an aggregation function such as mean, max,…
```

```python
monthly_max.index.freq
```

```python
NO2.resample("D").mean().plot(style = "-*",figsize = (10,5))
```
